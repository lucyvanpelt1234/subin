```python
#problem 6

class Flower:
    def __init__(self, name, color, blooming_time):
        self.name = name
        self.color = color
        self.blooming_time = blooming_time
        print("the name of the flower is", name)
        print("the color of the flower is", color)
        print("the blooming time of the flower is", blooming_time)
    
    def bloom(self):
        print(f"[{self.name}] can bloom")
    def grow(self):
        print(f"[{self.name}] can grow")
    def wither(self):
        print(f"[{self.name}] can wither")

rose = Flower("rose", "red", "may")
rose.bloom()
rose.grow()
rose.wither()
daisy = Flower("daisy", "white", "march")
daisy.bloom()
daisy.grow()
daisy.wither()
marigold = Flower("marigold", "red", "june")
marigold.bloom()
marigold.grow()
marigold.wither()

```
